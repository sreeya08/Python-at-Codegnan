```python
#empty tuple
t1 = ()
t2 = tuple()
print(type(t1),type(t2))
```

```python
#single element representation
t1 = (10)
t2 = (10,)
l1 = [10]
print(type(t1),type(t2),type(l1))
```

```python
#read tuple integer from user
t = tuple(map(int,input().split()))
print(t)
```

```python
#immutable 
t = (1,2,3,[10,20,30],"Codegna")
t[2] = 100
```

```python
#immutable 
# if a tuple contains list elements in it we can add elements in list because tuple doesnt change its position and length even elements are added
t = (1,2,3,[10,20,30],"Codegna")
print(t)
print(len(t))
t[3].append(100)
print(t)
print(len(t))
```

```python
#tuple operations
#1.tuple concatenation
#2. tuple repetition
t1 = (1,2,3)
t2 = (4,5,6)
print(t1 + t2)
print(t1 * 3)
print((0,) * 10)
```

```python
#tuple methods
# 1.count(item)
# 2.index(item)
t1 = (1,2,3)
print(t1.count(10))
print(t1.index(3))
print(t1.index(10))
```
