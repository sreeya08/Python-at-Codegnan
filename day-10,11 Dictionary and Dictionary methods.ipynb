#### Dictionary methods
- value getting from dictionary
    - get(key, default_val) - it returns specified keys value if it is exists in dictionary otherwise default_value
    - update dictionary items
    - update(new_dict) - it updates Dictionary items with new_dict items if keys already exists otherwise add's new item
- Get keys,values,Items from Dictionary
    - keys() - it returns list of dictionary keys
    - values() - it returns list of dictionary values
    - items() - it returns list of tuple of key-value pairs
- Delete items from dictionary
    - pop(key) - it deletes the item and returns the value of specified key,otherwise 'keyerror'
    - popitem() - it delete and returns the last item from dictionary

```python
#get
details = {'name': 'Sreeya', 'age': -20, 'course': 'Python'}
#details['Location'] # key error
print(details.get('name',"Name not exists"))
print(details.get('location'))
print(details.get('location','India'))
```

```python
#update(new_dict)
details = {'name': 'Sreeya', 'age': 20, 'course': 'Python'}
print(details)
details.update({"age":22,"course":"sql"})
print(details)
```

```python
#keys(),values(),items()
details = {'name': 'Sreeya', 'age': 20, 'course': 'Python'}
dict_keys = details.keys()
print(type(dict_keys))
print(dict_keys)
print(details.values())
print(details.items())
```

```python
details =  {'name': 'Sreeya', 'age': 30, 'course': 'Python'}
# print(details.pop('Location')) # key error
print(details.pop('age'))
print(details)
print(details.popitem())
print(details)
```
