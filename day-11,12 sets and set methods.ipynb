##### set operations or set methods
- Adding elemnts to set
    - add(ele) - it add the element to set,if element not exists in set, otherwise nothing
    - update(new_set) - it add all new set element to existing set
    - removing element from sets
    - remove(ele) - it remove the element form set if it exists otherwise 'KeyError'
    - discard(ele) - it removes the element from set if it exists otherwise nothing
    - set operations
    - union(new_sets) - it returns the all elements from the all sets
    - intersection(new_sets) - it returns the common elements from all sets
    - difference(new_sets) - it returns the set elements those are not exists in new_sets
    - symmetric_difference(new_sets) - it returns the non-common

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
print(s1.union(s2))
print(s1.intersection(s2))
print(s1.difference(s2))
print(s1.symmetric_difference(s2))
```

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
union = s1.union(s2)
print(union)
intersection = s1.intersection(s2)
print(intersection)
print(s1,s2)
#s1.intersection_update(s2)
difference = s1.difference(s2)
print(difference)
symmetric = s1.symmetric_difference(s2)
print(symmetric)
```

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
union = s1 | s2
print(union)
intersection = s1 & s2
print(intersection)
print(s1,s2)
#s1.intersection_update(s2)
difference = s1 - s2
print(difference)
symmetric = s1 ^ s2
print(symmetric)
```

```python
#issubset(), issuperset(), isdisjoint()
s1 = {1,2,3}
s2 = {1,2,3,4,5}
s3 = {7}
print(s2.issubset(s1))
print(s1.issubset(s2))
print(set().issubset(s2))
print(s2.issuperset(s1))
print(s1.issubset(s2))
print(s1.isdisjoint(s3))
```
