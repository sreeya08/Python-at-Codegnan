```python
#if target exist in list, then find what is positive or negative
arr = list(map(int,input().split()))
target = int(input())
if target in arr:
    if target >= 0:
        print("positive")
    else:
        print("negative")
else:
    print("target not in list")
```

```python
arr = list(map(int,input().split()))
target = int(input())
arr_length = len(arr)
if arr_length > target:
    print("more elements in tuple")
elif arr_length < target:
    print("less elements in tuple")
else:
    print("equal number of elements in tuple")
```
