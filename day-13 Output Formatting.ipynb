```python
# using modulo operator
#- %d - integers
#- %f - float
#- %s - string
name = "sreeya"
age = 20
marks = [90,79,59,87,56,87]
print("my name is %s and i am a %d years old and marks %s"%(name, age, marks))
```

```python
#using dot format
name = "sreeya"
age = 20
marks = [90,79,59,87,56,87]
print("my name is {} and i am a {} years old and marks {}".format(name, age, marks))
```

```python
#using F-format method
name = "sreeya"
age = 20
marks = [90,79,59,87,56,87]
print(f"my name is {name} and i am a {age} years old and marks {marks}")
```

```python
# 
per = 95.254685
name = "sreeya"
age = 20
print(f"my name is {name} and i am a {age} years old and percentage: {per:.2f}")
```

```python
# n1 = 10,n2 = 11,n3 = 12 print 10 is even 11 is odd 12 is even
n1 = int(input())
n2 = int(input())
n3 = int(input())
if n1 % 2 == 0:
    print(f"{n1} is even")
else:
    print(f"{n1} is odd")
if n2 % 2 == 0:
    print(f"{n2} is even")
else:
    print(f"{n2} is odd")
if n3 % 2 == 0:
    print(f"{n3} is even")
else:
    print(f"{n3} is odd")
```
