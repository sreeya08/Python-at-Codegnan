```python
# print 1 to N numbers
n = int(input())
for num in range(1, n+1):
    print(num)
else:
    print("Program done")
```

```python
# print 1 to N numbers
n = int(input())
for num in range(n, 0, -1):
    print(num)
else:
    print("Program done")
```

```python
#find the even numbers between 1-N and return it as list
n = int(input())
res = list()
for num in range(1, n+1):
    if num % 2 == 0:
        res.append(num)
print(res)
```

```python
#find the even numbers between 1-N and return it as list
n = int(input())
res = list()
for num in range(2, n+1, 2):
    res.append(num)
print(res)
```

```python
#count vowels and consonants in a given string
s = input().lower()
s = s.replace(" ","")
print(s)
vowel = 0
consonant = 0
for char in s:
    if char in 'aeiou':
        vowel += 1
    else:
        consonant += 1
print("Vowels:",vowel)
print("Consonants:",consonant)
print(vowel,consonant)
```

```python
#check whether target exists in list or not if it is exist return index otherwise -1
#loop control statements
arr = list(map(int,input().split()))
target = int(input())
n = len(arr)
for i in range(n):
    if target == arr[i]:
        print(i)
        break
else:
    print(-1)
```
