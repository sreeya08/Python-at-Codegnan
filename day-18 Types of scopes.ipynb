```python
#trying to update global variable in local scope
x = 10
def outer():
    y = 20
    x = x + y
    print("Local scope:", x)
    def inner():
        z = 30
        print("Non local scope:", x + y + z)
    inner()
outer()
print("global scope:", x)
```

```python
#trying to update global variable in local scope by using 'global' keyword
x = 10
def outer():
    global x
    y = 20
    x = x + y
    print("Local scope:", x)
    def inner():
        z = 30
        print("Non local scope:", x + y + z)
    inner()
outer()
print("global scope:", x)
```

```python
#trying to update global variable in local scope by using 'global' keyword
#updating local scope varible in nonlocal scope by using 'nonlocal' keyword
x = 10 #global variable
def outer():
    global x
    y = 20 #local variable
    x = x + y
    def inner():
        nonlocal y
        z = 30 #non local variable
        y = y + z + x
        print("Non local scope:", y)
    inner()
    print("Local scope:", y)
outer()
print("global scope:", x)
```

```python
#updating counter
def counter(x):
    print("-----------function execution start----------------")
    print(f"before updating x = {x} value reference {id(x)}")
    x = x + 1
    print(f"after updating x = {x} value reference {id(x)}")
    print("-----------function execution completed------------")
a = 10
print(f"before function execution {a} value reference:{id(a)}")
counter(a)
print(f"after function execution {a} value reference:{id(a)}")
```

```python
#updating counter
def counter(x):
    print("-----------function execution start----------------")
    print(f"before updating x = {x} value reference {id(x)}")
    x.append(10)
    print(f"after updating x = {x} value reference {id(x)}")
    print("-----------function execution completed------------")
a = [1,2,3]
print(f"before function execution {a} value reference:{id(a)}")
counter(a)
print(f"after function execution {a} value reference:{id(a)}")
```
