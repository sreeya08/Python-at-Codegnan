```python
#sum of two numbers
#function 
#->int returns function value type
def addition(a:int,b:int)->int:
    return a + b
#function calling
x,y = map(int,input().split())
res = addition(x, y)
print(res)
```

```python
#sum of three numbers
#function 
#->int returns function value type
def sumofthreenumbers(x:int,y:int, z:int)->int:
    return x + y + z
#function calling
a,b,c = map(int,input().split())
res = sumofthreenumbers(a, b, c)
print(res)
```

```python
#types of arguments
#1.positional arguments: the arguments
def subraction(x:int,y:int)->int:
    return x - y
ans = subraction(10, 5)
print(ans)
print(subraction(5,10))
```

```python
#types of arguments
#1.keyword arguments : the arguments are passed to function calling along with parameter
def subraction(x:int,y:int)->int:
    return x - y
ans = subraction(x = 10,y = 5)
print(ans)
print(subraction(y = 5,x = 10))
```

```python
def student(name:str, age:int, **kwargs):
    print(f"Name: {name}, Age: {age}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
student(name="sreeya", age=21, location="Hyderabad", education="B.Tech")
```

```python
#counter using lambda
counter = lambda x: x + 1
print(counter(2))
print(counter(counter(10)))
```

```python
even_or_odd = lambda x: f"{x} is Even" if x%2==0 else f"{x} is odd"
print(even_or_odd(10))
print(even_or_odd(11))
```

```python
#check whether the string contains a character or not 
word = "helloworld"
char = 'w'
check_char = lambda word,char: f"{char} exists in {word}" if char in word  else f"{char} not exists in {word}"
print(check_char(word,char))
```
