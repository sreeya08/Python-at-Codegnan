```python
def factorial(x):
    if x==1 or x==0:
        return 1
    return x * factorial(x-1)
factorial(5)
```

```python
#print numbers from 1 to n using recursion
def fun(n):
    if n == 1:
        print(1)
        return
    fun(n-1)
    print(n)
fun(10)
```

```python
#find even or odd using #indirect recursion
def is_even(n):
    if n==0:
        return True
    return is_odd(n-1)
def is_odd(n):
    if n==0:
        return False
    return is_even(n-1)
n = int(input()) 
if is_odd(n):
    print(f"{n} is odd")
else:
    print(f"{n} is even")
```
