#### math functions
-len(iterable) : it returns the length of iterable
-min(iterable) : it returns the smallest value from iterable
-max(iterable) : it returns largest value from iterable
-sum(iterable) : it returns the sum of all elements in iterable

```python
arr1 = [1,2,3,4,5]
arr2 = ['abc','abd','acd']
print(len(arr1),len(arr2))
print(min(arr1),min(arr2))
print(max(arr1),max(arr2))
print(sum(arr1))
```

```python
#abs(val) : it returns the absolute values
print(abs(10))
print(abs(-9))
```

```python
#round(val, pos):
val = 10.2545852458
print(round(val, 2))
```

```python
#enumerate(iterable): it returns the list of tuple of index and value pairs
#map(fun, iterable): it applies the functionality on every element
#filter(fun, iterable): it filter the all iterable element,it returns the elements which is satisfy the condition
#zip(iterable1, iterable2,...):
lst1 = [10,20,30,40]
lst2 = ['a','b','c']
print(enumerate(lst1))
print(list(enumerate(lst1)))
```

```python
# return the target element position in list
lst = list(map(int,input().split()))
target = int(input())
for ind, val in enumerate(lst):
    if val == target:
        print(ind)
        break
else:
    print(-1)
    
```

```python
#map()
def square(x):
    return x**2
lst = [1,2,3,4,5]
#map(square, lst) - it gives object value
print(list(map(square, lst)))
```

```python
#map using lambda
lst = [1,2,3,4,5]
print(list(map(lambda x:x**2, lst)))
```

```python
#return the even number from list
def evenorodd(x):
    if x%2 == 0:
        return x
lst = [12,33,43,54,65,24]
print(list(filter(evenorodd, lst)))
```

```python
#return the even number from list with map
def evenorodd(x):
    if x%2 == 0:
        return x
lst = [12,33,43,54,65,24]
print(list(map(evenorodd, lst)))
```

```python
#return the even number from list with lambda
lst = [12,33,43,54,65,24]
print(list(filter(lambda x:x%2==0, lst)))
```

```python
#return the even number or odd number from list with lambda
lst = [12,33,43,54,65,24]
print(list(map(lambda x:"even" if x%2==0 else "odd", lst))) #ternary operator
print(list(map(lambda x:f"{x} - even" if x%2==0 else f"{x} - odd", lst))) 
```

```python
#zip(iterable1, iterable2,...):
lst1 = [10,20,30,40]
lst2 = ['a','b','c']
print(list(zip(lst1, lst2)))
```

```python
#zip(iterable1, iterable2,...): returns the list of tuple packing of all index wise elements
telugu = [76,86,77,100]
english = [65,76,87]
hindi = [76,87,98]
res = list(zip(telugu,english,hindi))
print(list(enumerate(res)))
```

```python
#all(iterable): it returns True if all are non zero
#any(iterable): it returns true if any one element in iterable is non zero 
print(all([1,2,3,4]))
print(all([1,2,3,4,0]))
print(any([1,23,4,0,0,0,0]))
print(any([0,0,0,0]))
```
