#### string methods
#.Case Conversion methods
    -lower() - it will convert entire string into lower case
    -upper() - it will convert entire string into upper case
    -title() - it converts first character of each word into upper case and remaining into lower case
    -swapcase() - it will convert lower case letter into upper case letter and upper case letter into lower case
    -capitalize() - it converts first character into upper case and remaining into lower case

```python
s = "Forever and Ever"
l = s.lower()
print(l)
print(s.upper())
print(s.lower())
print(s.title())
print(s.swapcase())
print(s.capitalize())
```

#find substring index position
.index(val) - It returns the first occurence index of substring,otherwise 'valueError'
.#find(val) - if substring exists,it return the first occurence index of substring, otherwise -1

```python
s = "Forever and EveR"
print(s.index('e'))
print(s.find('e'))
print(s.find('z'))
print(s.index('z'))
```

```python
s = "Forever and Ever"
print(s.index('er'))
print(s.index('er', 8)) # it start checking from 8th position
```

#### whitespace or pattern removing
-strip() - It removes the whitespace to both side of string
-lstrip() - It removes the left side whitespace to the string
-rstrip()  - It removes the right side whitespace to the string

```python
s = "       Codegnan             "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s = "@@@@@Code@@@gnan@@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

## count the frequency of substring
- count(sub_string) - it returns the total count of non overlapping substring

```python
s = "ABCACACcccc"
print(s.count('C'))
print(s.count('CAC'))
#count the all 'c'(upper + lower)
print(s.count('c') + s.count('C'))
str1 = s.lower()
print(str1.count('c'))
```

```python
# split() - it splits the entire string into list of sub-string, 
#         by default it splits at space
s = "I am a PyThon batch student"
print(s.split())
print(s.split("a"))
print(s.split("t"))
print(s.lower().split('t'))
```

```python
#"sep".join(iterable) - It combines the all items in iterable with seperator
words = ['I', 'am', 'a', 'PyThon', 'batch', 'student']
print("".join(words))
print(" ".join(words))
print("_".join(words))
print("-".join(words))
```

```python
# replace(old, new) - It replace the old string with new string
s = "Python"
print(s.replace('P','J'))
print("AABBAAABBAABBCCC".replace("A","Z",5)) #it can replaces at upto 5 places
print("AABBAAABBAABBCCC".replace("A","Z",100))
print("AAASSSBBBIII".replace("S","C",10))
```

#### string checking methods
- islower() - it returns True if Entire string is in lower case
- isupper() - it returns False if Entire string is in upper case
- istitle() - it returns True if Entire string is in title format
- iscapitalize() - it returns True if Entire string is in capitalize format
- isdigit() - it returns True if entire string contains all digits
- isalpha() - it returns True if entire string contains alphabets
- isalnum() - it returns True if entire string contains either alphabets or digits
- startswith(sub_string) - it returns True if string starts with substring
- endswith(sub_string) - it returns True if string ends with substring

```python
print("code".islower())
print("Code".islower())
print("code123".islower())
print("CODE".isupper())
print("Code".isupper())
print("CODE123".isupper())
print("1234".isdigit())
print("1234a".isdigit())
print("1234a".isalnum())
print("1234a".isalpha())
print("Codegnan".startswith("Co"))
print("Codegnan".endswith("an"))
```

```python
#string comparision - lexicographical order(left to right)
s1 = "abc"
s2 = "adc"
s3 = "ABC"
print(s1 > s2)
print(s1 > s3)
```
