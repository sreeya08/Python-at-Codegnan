#### List Methods
- items adding to list
  - append(item) - it adds element at end of list
  - extend(new_list) - it adds all new_list elements at end of the list
  - insert(pos, item) - it add the item at specified position


```python
l = [1,2,3]
l.append(10)
print(l)
l.extend([5,6,7])
print(l)
l.insert(2,100)
print(l)
l.insert(100,2000)
print(l)
l.insert(99,1000)
print(l)
```

#### delete itmes from list
- remove(item) - It removes the specified item from the list if it exists, otherwise `ValueError`
- pop(pos) - It remove and returns the value from the list for the specified position, otherwise `IndexError`. By default it will selects last postion.
- clear() - It removes the all elements from list 
- del - It remove the either items of list from memory

```python
l = [10,2,3,4,10,2]
l.remove(10)
print(l)
val = l.pop()
print(val, 1)
val2 = l.pop(1)
print(val2,l)
#l.pop(100) gives index error
l.clear()
print(l)
del l
#print(l) - name error
```
